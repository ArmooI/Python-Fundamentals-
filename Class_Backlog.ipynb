# definition of http methods in  API
- http method: defines how a 
- get: retrieve data 
- post: send data to the server
- put: update data 
- delete: remove data / resources

```python
# API url and key

base_url = 'https://newsapi.org/v2/top-headlines'

api_key = 'd8c7db71fc254cb5a880817a2d510821'

# setting parameters for the news headlines needed

params = {
    "apikey": api_key, # this uses the API url and key defined above
    "country": "US", # this specifies the country from which the news headlines should come from 
    "pageSize": 100, # this specifies the number of pages the news headlines should be 
    "category": "technology" # this is the industry or area from which the news should come from 
}

# making request - This is the result of the above parameters specified with the API above 
response = requests.get(base_url, params = params)

```

```python
# checking if the request is successful - Using error checks 

if response.status_code == 200:
    data = response.json() # if the abov status code is 200, this will convert the data to dictionary format from json
    articles = data['articles']

else:
    print(f"error: {response.status_code}")
```

```python
df = pd.DataFrame(articles)

(df)
```

```python
# downloading the data retrieved as .csv file 

df.to_csv('technews_api.csv', index = False) # the false index ignores the numberring in the file
```
